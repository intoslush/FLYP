```python
import os
import pandas as pd
import argparse


templates_cal = [
    lambda c: f'a photo of a {c}.', lambda c: f'a painting of a {c}.',
    lambda c: f'a plastic {c}.', lambda c: f'a sculpture of a {c}.',
    lambda c: f'a sketch of a {c}.', lambda c: f'a tattoo of a {c}.',
    lambda c: f'a toy {c}.', lambda c: f'a rendition of a {c}.',
    lambda c: f'a embroidered {c}.', lambda c: f'a cartoon {c}.',
    lambda c: f'a {c} in a video game.', lambda c: f'a plushie {c}.',
    lambda c: f'a origami {c}.', lambda c: f'art of a {c}.',
    lambda c: f'graffiti of a {c}.', lambda c: f'a drawing of a {c}.',
    lambda c: f'a doodle of a {c}.', lambda c: f'a photo of the {c}.',
    lambda c: f'a painting of the {c}.', lambda c: f'the plastic {c}.',
    lambda c: f'a sculpture of the {c}.', lambda c: f'a sketch of the {c}.',
    lambda c: f'a tattoo of the {c}.', lambda c: f'the toy {c}.',
    lambda c: f'a rendition of the {c}.', lambda c: f'the embroidered {c}.',
    lambda c: f'the cartoon {c}.', lambda c: f'the {c} in a video game.',
    lambda c: f'the plushie {c}.', lambda c: f'the origami {c}.',
    lambda c: f'art of the {c}.', lambda c: f'graffiti of the {c}.',
    lambda c: f'a drawing of the {c}.', lambda c: f'a doodle of the {c}.'
]
templates_sample = [
    lambda c: f'a photo of a {c}.', 
]
templates_food = [
    lambda c: f'a photo of a {c}, a type of food.', 
    lambda c: f'a photo of the {c}, a type of food.', 
]
templates_animal = [
    lambda c: f'a photo of a {c}, a type of animal.', 
    lambda c: f'a photo of the {c}, a type of animal.', 
]
templates_dog = [
    lambda c: f'a photo of a {c}, a type of dog.', 
    lambda c: f'a photo of the {c}, a type of dog.',
]
def main(args):

    count_cal=0
    count_dog=0
    count_ani=0
    count_food=0
    print(args.data_dir)
    classes_in_dir = sorted(next(os.walk(args.data_dir))[1])#把目录里面的类排序然后和上面的替换,之后再根据模板制作csv
    print(args.data_dir,len(classes_in_dir))
    
    

    with open(args.save_file, 'w') as f:
        f.write('title\tfilepath\n')
        for i, dir_name in enumerate(classes_in_dir):
            ori_path=dir_name
            if len(dir_name.split("&"))!=2:
                print(dir_name)

            dir_name,set_name=dir_name.split("&")
            dir_name=dir_name.replace("_"," ")
            if set_name=="Thu-dog":
                dir_name=dir_name.title()
                templates=templates_dog
                count_dog+=1
            elif set_name=="Food-101":
                dir_name.lower()
                templates=templates_food
                count_food+=1
            elif set_name=="Animal":
                dir_name.lower()
                templates=templates_animal
                count_ani+=1
            elif set_name=="Caltech-101":
                templates=templates_cal
                count_cal+=1
            else:
                print(dir_name,"!!!!!!!!!!!")
                templates=templates_sample
                
            for file in os.listdir(os.path.join(args.data_dir, ori_path)):
                full_path = os.path.join(args.data_dir, ori_path, file)
                for template in templates:
                        f.write(f'{template(dir_name)}\t{os.path.join("./datasets",full_path[2:])}\n')
                
    print("count",count_cal,"len(classes_in_dir)",len(classes_in_dir))
    print("count_cal",count_cal,"count_food",count_food,"count_ani",count_ani,\
          "count_dog",count_dog,"\ntotal",count_cal+count_food+count_ani+count_dog)
    # assert count_cal == 101, 'number of classes mismatch'
    
    assert len(classes_in_dir) == 374, 'number of classes mismatch'

from types import SimpleNamespace

args={
    "save_dir":'./csv',
    "data_dir":'./data',
    "data_name":'compitition',
    'type':'train'
}
args = SimpleNamespace(**args)
args.data_dir = os.path.join(args.data_dir, args.data_name, args.type)
args.save_file = os.path.join(args.save_dir, args.data_name,
                                f'{args.type}.csv')
os.makedirs(os.path.join(args.save_dir, args.data_name), exist_ok=True)
main(args)
```
